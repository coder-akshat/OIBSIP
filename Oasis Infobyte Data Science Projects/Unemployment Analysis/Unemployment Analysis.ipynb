```python
#Renaming the columns headings of 1st dataset
df.columns=["State","Date","Frequency","Estimated Unemployment Rate","Estimated Employed","Estimated Labour Participation","Area"]
df.head()
```

```python
#Renaming the columns headings of 2nd dataset
df_2020.columns=["State","Date","Frequency","Estimated Unemployment Rate","Estimated Employed","Estimated Labour Participation","Region","Langitude","Longitude"]
df.head()
```

```python
#State with Highest Unemployment and lowest Employment
h_state=df["State"].value_counts().idxmax()
l_state=df["State"].value_counts().idxmin()
print("State with highest unemployment:",h_state)
print("State with lowest unemployment:",l_state)
```

```python
#Adding new month and month_int columns to 1st dataset
df["Date"] = pd.to_datetime(df["Date"],dayfirst = True)
df["Month_Int"] = df["Date"].dt.month
df["Month"] = df["Month_Int"].apply(lambda x: calendar.month_abbr[x])
```

```python
#Adding new month and month_int columns to 2nd dataset
df_2020["Date"] = pd.to_datetime(df_2020["Date"],dayfirst = True)
df_2020["Month_Int"] = df_2020["Date"].dt.month
df_2020["Month"] = df_2020["Month_Int"].apply(lambda x: calendar.month_abbr[x])
```

```python
#Month with highest unemployment and lowest unemployment for 1st dataset
h_month=df["Month"].value_counts().idxmax()
l_month=df["Month"].value_counts().idxmin()
print("Month with highest unemployment:",h_month)
print("Month with lowest unemployment:",l_month)
```

```python
#Month with highest unemployment and lowest unemployment for 2nd dataset
h_month_2020=df_2020["Month"].value_counts().idxmax()
l_month_2020=df_2020["Month"].value_counts().idxmin()
print("Month with highest unemployment:",h_month_2020)
print("Month with lowest unemployment:",l_month_2020)
```

```python
#Retreiving first 5 rows of 2nd dataset
df_2020.head()
```

```python
#Deleting Frequency and month_int columns from 1st dataset
df.drop(columns=["Frequency","Month_Int"])
```

```python
#Deleting Frequency and month_int columns from 2nd dataset
df_2020.drop(columns=["Frequency","Month_Int"])
```

```python
#Top 10 states with highest unemployment
df1 = df[["State","Estimated Unemployment Rate"]].groupby("State").sum().sort_values("Estimated Unemployment Rate", ascending=False)
df1.head(10)
```

```python
#Top 10 states with highest unemployment
df1_2020 = df_2020[["State","Estimated Unemployment Rate"]].groupby("State").sum().sort_values("Estimated Unemployment Rate", ascending=False)
df1_2020.head(10)
```

```python
#Plotting top 10 states with highest unemployment using 1st dataset

fig = plt.figure()
ax0 = fig.add_subplot(1,2,1)
df1[:10].plot(kind="bar",color="red",figsize=(30,5),ax=ax0)
ax0.set_title("Top 10 states with highest unemployment")
ax0.set_xlabel("States")
ax0.set_ylabel("Number of people unemployment percentage")
```

```python
#Plotting top 10 states with highest unemployment using 2nd dataset

fig = plt.figure()
ax0 = fig.add_subplot(1,2,1)
df1_2020[:10].plot(kind="bar",color="blue",figsize=(30,5),ax=ax0)
ax0.set_title("Top 10 states with highest unemployment")
ax0.set_xlabel("States")
ax0.set_ylabel("Number of people unemployment percentage")
```

```python
#Top 10 months with highest unemployment
df2 = df[["Month","Estimated Unemployment Rate"]].groupby("Month").sum().sort_values("Estimated Unemployment Rate", ascending=False)
df2.head(10)
```

```python
#Top 10 months with highest unemployment
df2_2020 = df_2020[["Month","Estimated Unemployment Rate"]].groupby("Month").sum().sort_values("Estimated Unemployment Rate", ascending=False)
df2_2020.head(10)
```

```python
#Plotting top 10 months of highest unemployment using 1st dataset

fig = plt.figure()
ax0 = fig.add_subplot(1,2,1)
df2[:10].plot(kind="bar",color="yellow",figsize=(30,5),ax=ax0)
ax0.set_title("Top 10 months of highest unemployment")
ax0.set_xlabel("Month")
ax0.set_ylabel("Number of people unemployment percentage")
```

```python
#Plotting top 10 months of highest unemployment using 2nd dataset

fig = plt.figure()
ax0 = fig.add_subplot(1,2,1)
df2_2020[:10].plot(kind="bar",color="green",figsize=(30,5),ax=ax0)
ax0.set_title("Top 10 months of highest unemployment")
ax0.set_xlabel("Month")
ax0.set_ylabel("Number of people unemployment percentage")
```

```python
#Plotting labour participation rate and unemployment rates in each month for 1st dataset
df_EEE = df.groupby("Month")[["Estimated Unemployment Rate", "Estimated Employed", "Estimated Labour Participation"]].mean().reset_index()
df_EEE_melted = df_EEE.melt(id_vars="Month", var_name="Rate_Type", value_name="Rate")

subset_data = df_EEE_melted[df_EEE_melted["Rate_Type"].isin(["Estimated Unemployment Rate", "Estimated Labour Participation"])]
plt.figure(figsize=(10, 6))
sns.barplot(data=subset_data, x="Month", y="Rate", hue="Rate_Type", errorbar=None)  
plt.title("Unemployment Rate and Labour Participation Rate")
plt.xlabel("Month")
plt.ylabel("Rate")
plt.legend(title=None)
plt.ticklabel_format(style='plain', axis='y')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Plotting labour participation rate and unemployment rates in each month for 2nd dataset
df_2020_EEE = df_2020.groupby("Month")[["Estimated Unemployment Rate", "Estimated Employed", "Estimated Labour Participation"]].mean().reset_index()
df_2020_EEE_melted = df_2020_EEE.melt(id_vars="Month", var_name="Rate_Type", value_name="Rate")

subset_data_2020 = df_2020_EEE_melted[df_2020_EEE_melted["Rate_Type"].isin(["Estimated Unemployment Rate", "Estimated Labour Participation"])]
plt.figure(figsize=(10, 6))
sns.barplot(data=subset_data_2020, x="Month", y="Rate", hue="Rate_Type", errorbar=None)  
plt.title("Unemployment Rate and Labour Participation Rate")
plt.xlabel("Month")
plt.ylabel("Rate")
plt.legend(title=None)
plt.ticklabel_format(style='plain', axis='y')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
